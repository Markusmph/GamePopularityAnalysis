```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Import data

```python
data = pd.read_csv("DummyData.csv")
players_day = data['Number of players']
dates = data['Time(days)']
```

### Get average

```python
avg = sum(players_day) / len(players_day)
avg
```

### Plot

```python
plt.plot(players_day)
plt.ylabel('Number of players')
plt.xlabel('Time (hours)')
plt.show()
```

### Peak

```python
max_players_day = 0
for i in range(len(players_day)):
    if players_day[i] > max_players_day:
        max_players_day = players_day[i]
        max_players_day_date = dates[i]

max_players_day_date
```

### Week avg

```python
day = 0
current_week = 0
players_week = []
for num in players_day:
    if day < 7:
        current_week = current_week + num
        day = day + 1
    else:
        players_week.append(current_week / day)
        current_week = num
        day = 1
players_week
```

### Plot of week avg

```python
plt.plot(players_week)
plt.ylabel('Number of players')
plt.xlabel('Time (weeks)')
plt.show()
```
